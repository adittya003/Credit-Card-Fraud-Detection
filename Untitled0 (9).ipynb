Calculate age using dob

```python
df_train['dob'] = pd.to_datetime(df_train['dob'])
df_test['dob'] = pd.to_datetime(df_test['dob'])
```

```python
from datetime import date

def calculate_age(born):
  today = date.today()
  return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df_train['age'] = df_train['dob'].apply(calculate_age)
df_test['age'] = df_test['dob'].apply(calculate_age)
```

```python
def day_period(x):
    if x >=0 and x < 6: return "Night"
    elif x>= 6 and x <= 12: return "Morning"
    elif x> 12 and x <= 15: return "Afternoon"
    elif x> 15 and x <= 20: return "Evening"
    elif x> 20 and x <= 24: return "Night"


df_train["day_period"] = df_train["trans_date_trans_time"].dt.hour.apply(day_period)
df_test["day_period"] = df_test["trans_date_trans_time"].dt.hour.apply(day_period)
```

find the merchant most involved in fraud
there are totally 693 unique name of merchant

```python
merchant_groups1 = df_train.groupby('merchant')
merchant_groups2 = df_test.groupby('merchant')
fraud_percentage_dict1 = {}
fraud_percentage_dict2 = {}


for merchant, group in merchant_groups1:
    total_trans = group.shape[0]
    total_fraud_trans = group[group["is_fraud"] == 1].shape[0]
    fraud_percentage_dict1[merchant] = (total_fraud_trans / total_trans) * 100
for merchant, group in merchant_groups2:
    total_trans = group.shape[0]
    total_fraud_trans = group[group["is_fraud"] == 1].shape[0]
    fraud_percentage_dict2[merchant] = (total_fraud_trans / total_trans) * 100

# Step 2: Map the calculated percentages back to the DataFrame
df_train["fraud_merchant_pct"] = df_train["merchant"].map(fraud_percentage_dict1)
df_test["fraud_merchant_pct"] = df_test["merchant"].map(fraud_percentage_dict2)
```

Lets rank merchant based on fraud percentage

```python
sorted_fraud_percentage1 = sorted(fraud_percentage_dict1.items(), key=lambda x: x[1])
sorted_fraud_percentage2 = sorted(fraud_percentage_dict2.items(), key=lambda x: x[1])
# Create a dictionary for ordinal encoding
merchant_ordinal_dict1 = {merchant: rank + 1 for rank, (merchant, _) in enumerate(sorted_fraud_percentage1)}
merchant_ordinal_dict2 = {merchant: rank + 1 for rank, (merchant, _) in enumerate(sorted_fraud_percentage2)}
# Step 2: Map the ordinal encoding back to the DataFrame
df_train["merchant_encoded"] = df_train["merchant"].map(merchant_ordinal_dict1)
df_test["merchant_encoded"] = df_test["merchant"].map(merchant_ordinal_dict2)
```
